# Trying out K-means

Probably won't be better than gussing.

```python
import pickle as pkl
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
with open('../data/dataset0/pca_proc.pkl', 'rb') as fb:
    data, labels = pkl.load(fb)
```

```python
clusters = 2
model = KMeans(n_clusters=clusters, max_iter=1000000000).fit(data[:,:].T)
```

```python
tn, fp, fn, tp = confusion_matrix(model.labels_, labels).ravel()/1000
```

```python
tp/(tp+fn)
```

```python
tn/(tp+fn)
```

```python
tn/(tn+fn)
```

```python
groups = [data[0:3,model.labels_==grp] for grp in range(clusters)]
```

```python
threedee = plt.figure(figsize=(8,6)).gca(projection='3d')
for x,y,z in groups:
    threedee.scatter(x,y,z, alpha=0.8, marker='.')
```

```python
groups = [data[0:3,labels==grp] for grp in range(clusters)]
threedee = plt.figure(figsize=(8,6)).gca(projection='3d')
for x,y,z in groups:
    threedee.scatter(x,y,z, alpha=0.8, marker='.')
```
